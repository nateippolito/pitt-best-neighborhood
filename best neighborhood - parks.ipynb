## Introduction
In this project, our goal is to identify the "best" neighborhood in Pittsburgh using a data-driven approach. We believe that the quality of life in an urban environment is heavily influenced by access to green spaces and recreational facilities.

## My Process
I explored using total acreage of parks, but decided to focus on the availability of parks. A neighborhood with more parks offers more opportunities for exercise, relaxation, and community gathering.

## Metric
For this project, I defined the "best" neighborhood based on the metric of Outdoor Recreational Accessibility. I measured this by calculating the total number of parks per neighborhood. The dataset I used contains a list of all city-managed parks, including their names, types, maintenance responsibility, and geographic coordinates.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\paleg\Downloads\parks.csv")

# Count how many parks are in each neighborhood
park_counts = df['neighborhood'].value_counts()

# Show the top 10 neighborhoods
print("Top 10 Neighborhoods by Park Count:")
print(park_counts.head(10))

# Create a bar chart of the top 10
park_counts.head(10).plot(kind='bar', figsize=(12,6), color='teal')

# Add labels and a title
plt.title('Top 10 Pittsburgh Neighborhoods by Number of Parks')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.xticks(rotation=45) # Tilts the names so they are easier to read

plt.show()
```

### Narrative: Why This is the Best Neighborhood
Based on my data analysis, Squirrel Hill South is the best neighborhood in Pittsburgh. With a total of 8 parks, it significantly beats other neighborhoods like Beechview and South Side Slopes, which tied for second place with 5 parks each. This data suggests that Squirrel Hill South provides the highest level of geographic accessibility to public recreation. While many neighborhoods have 1 or 2 parks, the high concentration in Squirrel Hill South makes it the best neighborhood.
